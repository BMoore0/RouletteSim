### Roulette simulation
1) Ball can land on red or black
2) Payout is 2x the bet amount for choosing right color
3) Player can only play while they still have money

```python
%matplotlib inline
import random #random number generator
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np


#class for roullette wheel
class rouletteWheel:
   
    #simulate roulette ball landing spot. 
    # set odd numbers to red, even to black, from numbers 1 to 36
    def play(self):
        roll = random.randint(1, 37)
        
        if(roll % 2 == 1):
            color = "red"
        else:
            color = "black"
        return color
    
    def playNum(self):
        roll = random.randint(1, 37)
        return roll

#r = rouletteWheel()
#for i in range(5):
    #print (r.play())
    
    
```

```python
#create class for the player, with startmoney, bet amount, number of times betting
class player:
    def __init__(self, money, bet, times, color):
        self.money = money
        self.bet = bet
        self.times = times
        self.color = color
        self.history = np.zeros(times)
    
    #array of history of bets
    def getHistory(self):
        return self.history
    
    #play the roulette game, add to pot if colors match, lose money if wrong color
    def play(self, rouletteWheel):
        curCash = self.money
        broke = False #while player isnt broke, they can keep playing
        for i in range(self.times):
            if broke == False:
                spin = rouletteWheel.play()
                result = self.winOrLose(spin)
                curCash = self.currentCash(curCash, result)
                self.history[i] = curCash
                print(curCash)
                if curCash <= 0:
                    broke == True
                    return
    
            
     
    
    #if the spin lands on same color that player bet on, player wins
    #else, player loses
    def winOrLose(self, spin):
        if spin == self.color:
            result = 1
        else:
            result = -1
        return result
    #user winOrLose result to add or subtract bet amount from the total
    def currentCash(self, amount, result):
        if result == -1:
            amount -= self.bet
        elif result == 1:
            amount += self.bet
            
        return amount
            
    
    
        
wheel = rouletteWheel()
player = player(100, 5, 100, 'black')
player.play(wheel)
    
    
```

```python
cash = player.getHistory()
fig, ax = plt.subplots()
y = player.getHistory()
x = np.linspace(0, 100, 100)
ax.plot(x, y, color = 'green')
ax.set_xlabel("Number of bets")
ax.set_ylabel("Money")
```

```python
class playerNum:
    def __init__(self, money, bet, times, number):
        self.money = money
        self.bet = bet
        self.times = times
        self.number = number
        self.history = np.zeros(times)
    
    #array of history of bets
    def getHistory(self):
        return self.history
    
    #play the roulette game, add to pot if colors match, lose money if wrong color
    def play(self, rouletteWheel):
        curCash = self.money
        broke = False #while player isnt broke, they can keep playing
        for i in range(self.times):
            if broke == False:
                spin = rouletteWheel.playNum()
                result = self.winOrLoseNum(spin)
                curCash = self.currentCash(curCash, result)
                self.history[i] = curCash
                print(curCash)
                if curCash <= 0:
                    broke == True
                    return
    
            
     
    
    #if the spin lands on same color that player bet on, player wins
    #else, player loses
    def winOrLoseNum(self, spin):
        if spin == self.number:
            result = 1
        else:
            result = -1
        return result
    #user winOrLose result to add or subtract bet amount from the total
    def currentCash(self, amount, result):
        if result == -1:
            amount -= self.bet
        elif result == 1:
            amount += 36 * self.bet
            
        return amount
    
wheel1 = rouletteWheel()
player1 = playerNum(250, 5, 100, 34)
player1.play(wheel1)
```
